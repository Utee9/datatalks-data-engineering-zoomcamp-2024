```python
##### Question 5. Three biggest pick up Boroughs


trips_sep_18 = join[join["lpep_pickup_datetime"].dt.date == pd.to_datetime('2019-09-18').date()]
```

```python
######  Question 6. Largest tip


sep_astoria = join[(join['Zone']=='Astoria') &  (join["lpep_pickup_datetime"].dt.to_period('M') == '2019-09')]

sep_astoria.groupby('Zonedropoff')['tip_amount'].max().reset_index().sort_values('tip_amount', ascending=False)
```
